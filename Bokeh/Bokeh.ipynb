```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from sklearn.datasets import load_iris
import pandas as pd

output_notebook()  # Shows plot inline in notebook

# Load dataset
iris = load_iris()
flowers = pd.DataFrame(iris.data, columns=iris.feature_names)
flowers.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# Create figure
p = figure(title='test_flower')
p.xaxis.axis_label = "petal_length"
p.yaxis.axis_label = "petal_width"

# Use scatter instead of deprecated circle()
p.scatter(x=flowers['petal_length'], y=flowers['petal_width'], size=8)

# Show plot in notebook
show(p)
```

```python
output_file("test.html")
p = figure(title = 'test_flower')
p.xaxis.axis_label = "petal_length"
p.yaxis.axis_label = "petal_width"
# p.circle(flowers['petal_length'], flowers['petal_width']) #circle for scatter plot

p.scatter(x=flowers['petal_length'], y=flowers['petal_width'], size=8)
# p.circle(x=flowers['petal_length'], y=flowers['petal_width'], size=8)
show(p)
```

```python
output_file("test.html")
p = figure(title = 'test_flower')
p.xaxis.axis_label = "petal_length"
p.yaxis.axis_label = "petal_width"
p.line(flowers['petal_length'], flowers['petal_width']) #line for line chart
show(p)
```

```python
x = [1, 2, 3, 4]
y = [3, 8, 4, 7]

p = figure(title="line_chart")
p.line(x, y)
show(p)
```
