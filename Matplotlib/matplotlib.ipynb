```python
# Pie Chart: To understand the percent distribution of data
langs = ["C", "Python", "Java"]
students = [20, 100, 40]

plt.pie(students, labels = langs, autopct = "%1.1f%%")
plt.title("Pie chart")
plt.show()
```

```python
#Focus on one segment

explode = (0.0, 0.1, 0.1)
langs = ["C", "Python", "Java"]
students = [20, 100, 40]

plt.pie(students, labels = langs, autopct = "%1.1f%%", explode = explode)
plt.title("Pie chart")
plt.show()
```

```python
explode = (0.0, 0.1, 0.1)
langs = ["C", "Python", "Java"]
students = [20, 100, 40]

plt.pie(students, labels = langs, autopct = "%1.1f%%", explode = explode, shadow = True) #Shadow
plt.title("Pie chart")
plt.show()
```

```python
# Area Plot: Variables changed over time

# An area plot, also called a stacked area plot, is a type of data visualization that shows how
# various quantitative factors have changed over time while highlighting how they all work together to form the whole. 
# It's particularly good for illustrating how elements of a whole change over time or along a continuous dimension.
# A layered representation of the data is produced in an area plot by layering many variables on top of one another. 
# The cumulative influence of all levels provides insight into the overall trend. Each layer represents a separate category or variable. 
# To distinguish between categories, the space beneath each layer is frequently colored.

years = [2021, 2018, 2022, 2023, 2017]
category1 = [20, 25, 30, 28, 35]
category2 = [10, 15, 20, 18, 25]
category3 = [5, 8, 10, 9, 15]

plt.stackplot(years, category1, category2, category3, labels = ["category1", "category2", "category3",],
             colors = ["skyblue","lightgreen", "orange"])
plt.xlabel("years")
plt.ylabel("Values")
plt.title("Area plot")
plt.legend()
plt.show()
```
