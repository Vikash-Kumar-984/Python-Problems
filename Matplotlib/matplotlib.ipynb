```python
#line plot
#whenever you want to depict the relation between two variables in the form of trend
#mostly in x axis the date - time column is there
x = [1, 2, 3, 4, 5] #assume x is no of day
y = [5, 2, 7, 8, 1] #y is price of stock
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y)
plt.title("line chart to understand the syntax")
plt.xlabel("day interval as x points")
plt.ylabel("price as y points")
plt.show() #It will remove the metadata
```

```python
plt.plot(x, y, color = "red")
plt.title("line chart to understand the syntax")
plt.xlabel("day interval as x points")
plt.ylabel("price as y points")
plt.show()
```

```python
plt.plot(x, y, color = "red", marker = "o")
plt.title("line chart to understand the syntax")
plt.xlabel("day interval as x points")
plt.ylabel("price as y points")
plt.show()
```

```python
plt.plot(x, y, color = "red", marker = "o", linestyle = '--')
plt.title("line chart to understand the syntax")
plt.xlabel("day interval as x points")
plt.ylabel("price as y points")
plt.show()
```

```python
plt.plot(x, y, color = "red", marker = "o", linestyle = '--', linewidth = 4)
plt.title("line chart to understand the syntax")
plt.xlabel("day interval as x points")
plt.ylabel("price as y points")
plt.show()
```

```python
plt.plot(x, y, color = "red", marker = "o", linestyle = '--', linewidth = 4, markersize = 15)
plt.title("line chart to understand the syntax")
plt.xlabel("day interval as x points")
plt.ylabel("price as y points")
plt.show()
```

```python
plt.plot(x, y, color = "red", marker = "o", linestyle = '--', linewidth = 4, markersize = 15)
plt.title("line chart to understand the syntax")
plt.xlabel("day interval as x points")
plt.ylabel("price as y points")
plt.grid()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [5, 1, 7, 8, 2]
y2 = [7, 8, 3, 5, 1]

plt.plot(x, y1, color = "red", marker = "o", linestyle = "--", linewidth = 3, markersize = 6, label="line y1 as tata motors")
plt.plot(x, y2, color = "b", marker = "o", linestyle = "--", linewidth = 3, markersize = 6, label="line y2 as tata power")

plt.title("Multiple lines in one plot")

plt.xlabel("Days")
plt.ylabel("stock price")
plt.legend()

plt.show()
```

```python
x = np.linspace(0, 10, 100)
```

```python
plt.plot(x, np.sin(x), '--')
```

```python
plt.plot(x, np.sin(x), '-')
```

```python
plt.figure() #initialising a page to make plot
plt.subplot(1, 2, 1) #1 Row, 2 Column, Select 1
plt.plot(x, np.sin(x))

plt.subplot(1, 2, 2) #1 Row, 2 Column, Select 2
plt.plot(x, np.cos(x))

plt.show()
```
