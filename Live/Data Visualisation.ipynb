```python
import pandas as pd
import numpy as np
#!pip install seaborn
#!pip install matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("vo_comp.csv")
```

```python
df.head()
```

```python


# 1. Histogram - QUANTITYORDERED
plt.figure(figsize=(8,5))
plt.hist(df['QUANTITYORDERED'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Quantity Ordered')
plt.xlabel('Quantity Ordered')
plt.ylabel('Frequency')
plt.show()

# 2. Bar Chart - Average SALES by DEALSIZE
avg_sales = df.groupby('DEALSIZE')['Sales'].mean()
plt.figure(figsize=(8,5))
plt.bar(avg_sales.index, avg_sales.values, color='lightgreen')
plt.title('Average Sales by Deal Size')
plt.xlabel('Deal Size')
plt.ylabel('Average Sales')
plt.xticks(rotation=45)
plt.show()

# 3. Pie Chart - Order STATUS distribution
status_counts = df['DEALSIZE'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Order Status Distribution')
plt.axis('equal')
plt.show()

# 4. Line Chart - Total SALES over time
sales_over_time = df.groupby('ORDERDATETIME')['Sales'].sum()
plt.figure(figsize=(10,5))
plt.plot(sales_over_time.index, sales_over_time.values, marker='o', linestyle='-', color='purple')
plt.title('Total Sales Over Time')
plt.xlabel('Order Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.tight_layout()
plt.show()

# 5. Scatter Plot - QUANTITYORDERED vs SALES
plt.figure(figsize=(8,5))
plt.scatter(df['QUANTITYORDERED'], df['Sales'], color='orange', alpha=0.6)
plt.title('Quantity Ordered vs Sales')
plt.xlabel('Quantity Ordered')
plt.ylabel('Sales')
plt.show()

# 6. Box Plot - SALES by PRODUCTLINE
product_lines = df['PRODUCTLINE'].unique()
data_to_plot = [df[df['PRODUCTLINE'] == pl]['Sales'] for pl in product_lines]

plt.figure(figsize=(10,6))
plt.boxplot(data_to_plot, labels=product_lines)
plt.title('Sales Distribution by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Set style
sns.set(style='whitegrid')

# 1. Histogram - QUANTITYORDERED
plt.figure(figsize=(8,5))
sns.histplot(df['QUANTITYORDERED'], bins=20, kde=True, color='skyblue')
plt.title('Histogram of Quantity Ordered')
plt.xlabel('Quantity Ordered')
plt.ylabel('Frequency')
plt.show()

# 2. Bar Chart - Average SALES by DEALSIZE
plt.figure(figsize=(8,5))
sns.barplot(x='DEALSIZE', y='Sales', data=df, estimator='mean', palette='viridis')
plt.title('Average Sales by Deal Size')
plt.xlabel('Deal Size')
plt.ylabel('Average Sales')
plt.xticks(rotation=45)
plt.show()

# 3. Pie Chart - Use Matplotlib (Seaborn doesn't support pie charts natively)
status_counts = df['DEALSIZE'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Order Status Distribution')
plt.axis('equal')
plt.show()

# 4. Line Chart - Total SALES over time
sales_over_time = df.groupby('ORDERDATETIME')['Sales'].sum().reset_index()
plt.figure(figsize=(10,5))
sns.lineplot(data=sales_over_time, x='ORDERDATETIME', y='Sales', marker='o')
plt.title('Total Sales Over Time')
plt.xlabel('Order Date')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()

# 5. Scatter Plot - QUANTITYORDERED vs SALES
plt.figure(figsize=(8,5))
sns.scatterplot(x='QUANTITYORDERED', y='Sales', data=df, hue='DEALSIZE', alpha=0.7)
plt.title('Quantity Ordered vs Sales')
plt.xlabel('Quantity Ordered')
plt.ylabel('Sales')
plt.show()

# 6. Box Plot - SALES by PRODUCTLINE
plt.figure(figsize=(10,6))
sns.boxplot(x='PRODUCTLINE', y='Sales', data=df, palette='Set3')
plt.title('Sales Distribution by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
