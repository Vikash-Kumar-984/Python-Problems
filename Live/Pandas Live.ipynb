```python
#sales bucket - low sales [<= 4000], medium sales[> 4000 and less than or equal to 8000], high sales[> 8000]
```

```python
def myfunc(x):
    if x <= 4000:
        return "low sales"
    elif x <= 8000:
        return "medium sales"
    else:
        return "high sales"
```

```python
df["sales_bucket"] = df["Sales"].apply(myfunc)
```
